```python
%sql
CREATE OR REPLACE TABLE default.silver_movies_part
USING DELTA
PARTITIONED BY (release_year)
AS
SELECT * FROM default.silver_movies;
```

```python
%sql
OPTIMIZE default.silver_movies_part
ZORDER BY (studio);
```
