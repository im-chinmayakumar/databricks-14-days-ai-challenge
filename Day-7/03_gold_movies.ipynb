```python
print("This is Gold Notebook")

dbutils.widgets.text("run_date", "2026-01-15")
run_date = dbutils.widgets.get("run_date")
print("Gold running for date:", run_date)

silver_df = spark.table("default.silver_movies")

from pyspark.sql.functions import count, sum, avg, col

gold_studio_df = (
    silver_df
    .groupBy("studio")
    .agg(
        count("*").alias("movie_count"),
        sum("budget").alias("total_budget"),
        sum("revenue").alias("total_revenue"),
        avg(col("imdb_rating_clean")).alias("avg_imdb_rating")
    )
)

gold_studio_df.show(10)

gold_studio_df.write \
    .mode("overwrite") \
    .option("mergeSchema", "true") \
    .saveAsTable("default.gold_studio_metrics")
```
